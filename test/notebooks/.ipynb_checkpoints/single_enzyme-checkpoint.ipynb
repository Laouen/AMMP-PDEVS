```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv('./data/single_enzyme.csv')
display(df.head())
```

```python
plt.figure(figsize=(8,4))
ax = sns.lineplot(
    x='simulation_virtual_time',
    y='amount',
    hue='metabolite_id',
    data=df)

labels = ax.get_xticklabels()
total_ticks = len(labels)
steps = total_ticks // 4

for i,l in enumerate(labels):
    if i%steps == 0:
        l.set_rotation(90)
    else:
        l.set_visible(False)

#TODO: agrupar por compartimientos y poner colores correctos
        
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
